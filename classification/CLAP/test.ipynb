For `model.load_ckpt`, you can delete the `finetune_ckpt` parameter to load the original model.

```python
import laion_clap
import glob
import json
import torch
import numpy as np

device = torch.device('cuda')

# download https://drive.google.com/drive/folders/1scyH43eQAcrBz-5fAw44C6RNBhC3ejvX?usp=sharing and extract ./ESC50_1/test/0.tar to ./ESC50_1/test/
esc50_test_dir = './dataset/ESC-50/test/*/'
class_index_dict_path = './CLAP-main/class_labels/ESC50_class_labels_indices_space.json'

# Load the model
model = laion_clap.CLAP_Module(enable_fusion=False, device=device)
# Delete `finetune_ckpt` to load the model without finetuned
model.load_ckpt(ckpt="./Checkpoint/630k-audioset-best.pt")
# model.load_ckpt(ckpt="./Checkpoint/630k-audioset-best.pt", finetune_ckpt="./Checkpoint/pretrain_epoch_15_lp_epoch_latest.pt")

# Get the class index dict
class_index_dict = {v: k for v, k in json.load(open(class_index_dict_path)).items()}

# Get all the data
audio_files = sorted(glob.glob(esc50_test_dir + '**/*.flac', recursive=True))
json_files = sorted(glob.glob(esc50_test_dir + '**/*.json', recursive=True))
ground_truth_idx = [class_index_dict[json.load(open(jf))['tag'][0]] for jf in json_files]
```

```python
ground_truth_idx
```

```python
with torch.no_grad():
    all_texts = ["This is a sound of " + t for t in class_index_dict.keys()]
    text_embed = model.get_text_embedding(all_texts)
```

```python
batch_size = 64
```

```python
with torch.no_grad():
    audio_embed_list = []
    count = 0
    while (count < len(audio_files)):
        if count + batch_size > len(audio_files):
            temp_audio_files = audio_files[count:]
        else:
            temp_audio_files = audio_files[count:count+64]
        temp_audio_embed = model.get_audio_embedding_from_filelist(x=temp_audio_files)
        audio_embed_list.append(temp_audio_embed)
        count += batch_size
```

```python
audio_embed = np.concatenate(audio_embed_list)
audio_embed.shape
```

```python
text_embed.shape
```

```python
ground_truth = torch.tensor(ground_truth_idx).view(-1, 1)

ranking = torch.argsort(torch.tensor(audio_embed) @ torch.tensor(text_embed).t(), descending=True)
preds = torch.where(ranking == ground_truth)[1]
preds = preds.cpu().numpy()

metrics = {}
metrics[f"mean_rank"] = preds.mean() + 1
metrics[f"median_rank"] = np.floor(np.median(preds)) + 1
for k in [1, 5, 10]:
    metrics[f"R@{k}"] = np.mean(preds < k)
# map@10
metrics[f"mAP@10"] = np.mean(np.where(preds < 10, 1 / (preds + 1), 0.0))

print(
    f"Zeroshot Classification Results: "
    + "\t".join([f"{k}: {round(v, 4):.4f}" for k, v in metrics.items()])
)
```
